```python
# Dummy labels
# labels = tf.random.uniform((num_days, num_stocks), minval=0, maxval=num_classes, dtype=tf.int32)

labels = tf.convert_to_tensor(y_train.values)
val_labels = tf.convert_to_tensor(y_val.values)
test_labels = tf.convert_to_tensor(y_test.values)
```

```python
from tensorflow.keras import layers, models
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import csv
import datetime


tf.random.set_seed(13)


class MetricsCallback(tf.keras.callbacks.Callback):
    def __init__(self, X_train, y_train, X_val, y_val, file_path):
        self.X_train = X_train
        self.y_train = y_train
        self.X_val = X_val
        self.y_val = y_val
        self.file_path = file_path
        with open(self.file_path, 'a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["epoch", "accuracy", "precision", "recall", "f1", "val_accuracy", "val_precision", "val_recall", "val_f1"])

    def on_epoch_end(self, epoch, logs=None):
        X = self.X_train
        y = self.y_train

        predictions = self.model.predict(X)
        predictions = tf.argmax(predictions, axis=-1)
        accuracy = accuracy_score(y.numpy().flatten(), predictions.numpy().flatten())
        precision = precision_score(y.numpy().flatten(), predictions.numpy().flatten(), average='weighted', zero_division=1)
        recall = recall_score(y.numpy().flatten(), predictions.numpy().flatten(), average='weighted')
        f1 = f1_score(y.numpy().flatten(), predictions.numpy().flatten(), average='weighted')

        X = self.X_val
        y = self.y_val

        predictions = self.model.predict(X)
        predictions = tf.argmax(predictions, axis=-1)
        accuracy_val = accuracy_score(y.numpy().flatten(), predictions.numpy().flatten())
        precision_val = precision_score(y.numpy().flatten(), predictions.numpy().flatten(), average='weighted', zero_division=1)
        recall_val = recall_score(y.numpy().flatten(), predictions.numpy().flatten(), average='weighted')
        f1_val = f1_score(y.numpy().flatten(), predictions.numpy().flatten(), average='weighted')

        # Write metrics to file
        with open(self.file_path, 'a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow([epoch+1, accuracy, precision, recall, f1, accuracy_val, precision_val, recall_val, f1_val])

# Define the SimpleNN model
def SimpleNN(hidden_size, num_classes):
    model = models.Sequential([
        layers.Dense(hidden_size, activation='LeakyReLU'),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model

# Parameters
# input_size = 128
hidden_size = 64
num_classes = 3
learning_rate = 2e-5
num_epochs = 10
file_path = f'metrics.csv'  # File to store metrics
file_path_logger = f'training.log'

# Initialize model
model = SimpleNN(hidden_size, num_classes)

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])
```

```python
# Dummy data (assuming you have daily embeddings for 500 stocks)
# Replace this with your actual data
# Let's say your data is a 3D tensor of shape (num_days, num_stocks, embedding_size)
num_days = 293  # Example number of days
num_stocks = 10  # Example number of stocks
# daily_embeddings = tf.random.normal((num_days, num_stocks, input_size))


# Create and add the custom callback
metrics_callback = MetricsCallback(x_train_tensor, labels, x_val_tensor, val_labels, file_path)
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)
csv_logger = tf.keras.callbacks.CSVLogger(file_path_logger)

callbacks = [metrics_callback, early_stopping, csv_logger]

# Training loop
model.fit(x_train_tensor, labels, validation_data=(x_val_tensor, val_labels), epochs=num_epochs, callbacks=callbacks)

# Once trained, you can use this model to predict probabilities for each class for new daily embeddings
```
