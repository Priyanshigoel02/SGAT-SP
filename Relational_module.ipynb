```python
def clean (rel):
    rel['company1'] = rel['company1'].apply(lambda x: x.split("entity/")[-1])
    rel['company2'] = rel['company2'].apply(lambda x: x.split("entity/")[-1])
    rel['relation'] = rel['relation'].apply(lambda x: x.split("direct/")[-1])
    rel = rel.drop_duplicates()
    rel = rel.reset_index(drop=True)
    return rel
```

```python
dict2 = {
    "P155": "Follows" ,
    "P749": "Parent Organization",
    "P452": "Industry",
    "P1454": "Legal Form",
    "P1830": "Owner Of",
    "P1056": "Product or material produced",
    "P127": "Owned By",
    "P355": "Subsidiary",
    "P112": "Founded By",
    "P5009": "Complies with"
}
dict2
```

```python
dict
```

```python
rev_dict = {}
for key, values in dict.items():
  for value in values:
    if value not in rev_dict:
      rev_dict[value] = [key]
    else:
      rev_dict[value].append(key)
```

```python
rev_dict
```

```python
def replace_symbols(row, dict, column):
    if row[column] in dict:
        return dict[row[column]]
    else:
        return row[column]
relation = pd.DataFrame()
relation['company1'] = rel.apply(lambda row: replace_symbols(row, rev_dict, 'company1'), axis=1).explode().reset_index(drop=True)
relation['relation'] = rel.apply(lambda row: replace_symbols(row, dict2, 'relation'), axis=1).explode().reset_index(drop=True)
relation['company2'] = rel.apply(lambda row: replace_symbols(row, rev_dict, 'company2'), axis=1).explode().reset_index(drop=True)
relation
```

```python
rel
```

```python
#TO DO: add extra rows for all ticker symbols correponding to the Q values :DONE
first_order = pd.DataFrame(columns=['Q1', 'ticker1', 'relation','ticker2', 'Q2'])
cnt = 0
for index, row in rel.iterrows():
    l1 = rev_dict[row['company1']]
    l2 = rev_dict[row['company2']]
    r = dict2[row['relation']]
    for ticker1 in l1:
        for ticker2 in l2:
            cnt = cnt + 1
            # newrow = {'ticker1': ticker1, 'ticker2':ticker2, 'relation':r}
            # print (newrow)
            # first_order = first_order.append(newrow, ignore_index=True)
            if (ticker1 != ticker2):
                first_order.loc[len(first_order)] = [row['company1'], ticker1, r, ticker2, row['company2']]
print(cnt)
first_order
```

```python
first_order_final = first_order.drop(['Q1', 'Q2'], axis=1)
first_order_final = first_order_final.drop_duplicates()
first_order_final
```

```python
#second order relation:
#first-order-relations dataframe
rel2 = pd.read_csv(path + "/secondOrder_query.csv")
rel2
```

```python
rel2 = clean(rel2)
rel2
```

```python
grouped = rel2.groupby('relation')
dfs_dict = {key: group for key, group in grouped}
dfs_dict.keys()
```

```python
second_order_relations = [
    ('P452', 'P1454'),
    ('P749', 'P1830'),
    ('P452', 'P1056'),
    ('P127', 'P355'),
    ('P112', 'P112'),
    ('P5009', 'P5009'),
    ('P355', 'P1830'),
    ('P1830', 'P749')
]
```

```python
second_order_map = pd.DataFrame(columns=['company1', 'relation1', 'relation2','company2'])

for r1, r2 in second_order_relations:
    df1 = dfs_dict[r1]
    df2 = dfs_dict[r2]

    for index1, row1 in df1.iterrows():
        for index2, row2 in df2.iterrows():
            if row1['company2'] == row2['company2']:
                second_order_map.loc[len(second_order_map)] = [row1['company1'], row1['relation'], row2['relation'], row2['company1']]

second_order_map
```

```python
sec_order = pd.DataFrame(columns=['Q1', 'ticker1', 'relation1', 'relation2','ticker2', 'Q2'])
cnt = 0

for index, row in second_order_map.iterrows():
    l1 = rev_dict[row['company1']]
    l2 = rev_dict[row['company2']]
    r1 = dict2[row['relation1']]
    r2 = dict2[row['relation2']]
    for ticker1 in l1:
        for ticker2 in l2:
            cnt = cnt + 1
            # newrow = {'ticker1': ticker1, 'ticker2':ticker2, 'relation':r1}
            # print (newrow)
            # # first_order = first_order.append(newrow, ignore_index=True)
            if ticker1 != ticker2:
                sec_order.loc[len(sec_order)] = [row['company1'], ticker1, r1, r2, ticker2, row['company2']]
print(cnt)
sec_order
```

```python
sec_order_final = sec_order.drop(['Q1', 'Q2'], axis=1)
sec_order_final = sec_order_final.drop_duplicates()
sec_order_final
```

```python
result = pd.concat([first_order_final[['ticker1', 'ticker2']], sec_order_final[['ticker1', 'ticker2']]], ignore_index=True)
result = result.drop_duplicates()
result
```
